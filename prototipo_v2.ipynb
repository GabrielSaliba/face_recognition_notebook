# Reconhecimento Facial

## Carregando imagens

```python
from PIL import Image #manipulação de imagens
from pathlib import Path #path diretórios do projeto
import face_recognition #reconhecimento facial
import cv2 #manipulação e transformação de imagens em arrays
```

#### Carregando imagem em cinza

```python
image_name = "Jennifer_Aniston_0016.jpg"
path = Path().resolve()
image = cv2.imread(f"{path}/images/{image_name}")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(image))
```

#### Obtendo somente o rosto da imagem e salvando

```python
face = face_recognition.face_locations(image)[0]
face_capture = image[face[0]:face[2], face[3]:face[1]]
file_save_path = f"{path}/outputs/face_{image_name}"
cv2.imwrite(file_save_path, face_capture)
```

#### Carregando a imagem salva e encodando

```python
image = face_recognition.load_image_file(file_save_path)
image_encoded = face_recognition.face_encodings(image)[0]
display(image_encoded) #Exibindo encoded image
```

## Treinamento de imagens

#### Salvando image encoded no array de treinamento

```python
training_array = []
training_array.append(image_encoded)
```

#### Carregando imagem para teste de reconhecimento

```python
image_name = "Jennifer_Aniston_0015.jpg"
path = Path().resolve()
image = cv2.imread(f"{path}/images/{image_name}")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(image))
face = face_recognition.face_locations(image)[0]
face_capture = image[face[0]:face[2], face[3]:face[1]]
display(Image.fromarray(face_capture))
file_save_path = f"{path}/outputs/face_{image_name}"
cv2.imwrite(file_save_path, face_capture)
```

#### Encodando imagem de teste

```python
image = face_recognition.load_image_file(file_save_path)
image_encoded_test = face_recognition.face_encodings(image)[0]
display(image_encoded_test) #Exibindo encoded image
```

## Reconhecimento facial

#### Comparando imagem de teste com a(s) image(s) do array de treinamento

```python
result = face_recognition.compare_faces(training_array, image_encoded_test)
distance = face_recognition.face_distance(training_array, image_encoded_test)
```

```python
print(f"Pessoa reconhecida na foto? {result[0]}")
print(f"Distância dos rostos identificados: {distance[0]}")
```

#### Face-distance = valor de 0 à 1 identificando o valor de confiabilidade no reconhecimento facial. Valores abaixo de 0.5 são rostos próximos o suficiente para serem identificados como da mesma pessoa

#### Adicionando mais imagens para treinamento

```python
image_name = "Jennifer_Aniston_0013.jpg"
path = Path().resolve()
image = cv2.imread(f"{path}/images/{image_name}")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(image))
face = face_recognition.face_locations(image)[0]
face_capture = image[face[0]:face[2], face[3]:face[1]]
display(Image.fromarray(face_capture))
file_save_path = f"{path}/outputs/face_{image_name}"
cv2.imwrite(file_save_path, face_capture)
```

```python
image = face_recognition.load_image_file(file_save_path)
image_encoded = face_recognition.face_encodings(image)[0]
```

```python
training_array.append(image_encoded)
len(training_array)
```

```python
result = face_recognition.compare_faces(training_array, image_encoded_test)
distance = face_recognition.face_distance(training_array, image_encoded_test)
```

#### Pode-se notar que com mais imagens para treinamento, o modelo se torna mais preciso

```python
print(result)
print(distance)
```
